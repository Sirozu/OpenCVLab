```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def select_image():
    path = "A.jpg"#"cat.jpg" #fd.askopenfilename()
    if (len(path) > 0):
        return cv2.imread(path, 0)
```

```python
def fft(im):
    dft = cv2.dft(np.float32(im), flags = cv2.DFT_COMPLEX_OUTPUT)
    return np.fft.fftshift(dft)
```

```python
def magnitude(fft_res):
    return 20*np.log(cv2.magnitude(fft_res[:,:,0], fft_res[:,:,1])+0.0001)
```

```python
def apply_mask(dft_shift, mask):
    # apply mask and inverse DFT    
    fshift = dft_shift*mask   
    f_ishift = np.fft.ifftshift(fshift)   
    img_back = cv2.idft(f_ishift)   
    img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])   
    
    return img_back
```

```python
def show_ims(img, img_back):
    size = 12
    plt.figure(num = 1, figsize=(size, size))
    plt.subplot(121),plt.imshow(img, cmap = 'gray')   
    plt.title('Input Image'), plt.xticks([]), plt.yticks([])   
    
    plt.subplot(122),plt.imshow(img_back, cmap = 'gray')   
    plt.title('Result'), plt.xticks([]), plt.yticks([])   
    plt.show()
```

```python
D = lambda u, v: np.sqrt((u - M//2)*(u - M//2) + (v - N//2)*(v - N//2))
```

```python
HBatt = lambda u, v: 1/(1 + (D(u,v)/D0)**(2*n))
```

```python
Hauss = lambda u, v: np.exp(-D(u,v)*D(u,v)/(2*D0*D0))
```

```python
im = select_image()
M, N = im.shape    
```

```python
plt.imshow(im, 'gray')
```

```python
dft_shift = fft(im)
amp = magnitude(dft_shift)
```

```python
plt.subplot(121)
plt.imshow(im, 'gray')
plt.subplot(122)
plt.imshow(amp, 'gray')
plt.show()
```

```python
crow,ccol = M//2 , N//2 
h = 20

circle = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*h,2*h))

mask = np.zeros((M,N,2),np.uint8)    
mask[crow-h:crow+h, ccol-h:ccol+h,0] += circle
mask[crow-h:crow+h, ccol-h:ccol+h,1] += circle

plt.imshow(mask[:,:,0], 'gray')
```

```python
plt.imshow(magnitude(dft_shift * mask), 'gray')
```

```python
im_after = apply_mask(dft_shift, mask)   
show_ims(im, im_after)

res_dft = fft(im_after)
res_amp = magnitude(res_dft)
show_ims(amp, res_amp)
```

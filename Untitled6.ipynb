```python
import cv2

import numpy as np

import matplotlib.pyplot as plt
```

```python
def select_image():
    path = "lena4.jpg" #fd.askopenfilename()
    if (len(path) > 0):
        return cv2.imread(path)
```

```python
def normalize(image):
    maxB = image[...,0].max()
    maxG = image[...,1].max()
    maxR = image[...,2].max()
    print(image)     
    return np.uint8(image[:,:] * [255/maxB, 255/maxG, 255/maxR])
```

```python
def base_color(image, base_colr, targ_colr):
    blue = targ_colr[0]/(base_colr[0]+1)
    green = targ_colr[1]/(base_colr[1]+1)
    red = targ_colr[2]/(base_colr[2]+1)
    print(blue, green, red)
    return np.uint8(np.clip(image * [blue, green, red], 0, 255))
```

```python
def gray_world(image):
    redAv = np.average(image[:,:,2])
    greenAv = np.average(image[:,:,1])
    blueAv = np.average(image[:,:,0])
    
    avg = (redAv + greenAv + blueAv)/3
    
    return np.uint8(np.clip(image[:,:] * [avg/blueAv, avg/greenAv, avg/redAv], 0, 255))
```

```python
im1 = select_image()
balanced_im = gray_world(im1)
im2 = select_image()
basecol_im = base_color(im2, im2[10,10], [192, 255, 192])
```

```python
im01 = cv2.cvtColor(im1, cv2.COLOR_BGR2RGB)
im02 = cv2.cvtColor(balanced_im, cv2.COLOR_BGR2RGB)
im03 = cv2.cvtColor(basecol_im, cv2.COLOR_BGR2RGB)

size = 20
plt.figure(num = 1, figsize=(size, size))
plt.subplot(121),plt.imshow(im01)   
plt.title('Basic'), plt.xticks([]), plt.yticks([])   
    
plt.subplot(122),plt.imshow(im03)   
plt.title('Hier'), plt.xticks([]), plt.yticks([]) 
plt.show()
```

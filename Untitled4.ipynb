```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def select_image():
    path = "Skelet.png" #fd.askopenfilename()
    if (len(path) > 0):
        return cv2.imread(path, 0)
```

```python
im = select_image()
```

```python
plt.imshow(im, 'gray')
```

```python
#lap_im = cv2.convertScaleAbs( cv2.Laplacian(im, cv2.CV_16S, 3) )
lap_im = cv2.Laplacian(im, cv2.CV_16S, 3)
```

```python
plt.imshow(lap_im, 'gray')
```

```python
grad_x = cv2.convertScaleAbs( cv2.Sobel(im, cv2.CV_16S, 1, 0, 3) )
grad_y = cv2.convertScaleAbs( cv2.Sobel(im, cv2.CV_16S, 0, 1, 3) )

grad_im = grad_x + grad_y
```

```python
plt.imshow(grad_im, 'gray')
```

```python
grad_blur = cv2.blur(grad_im, (5, 5))
```

```python
plt.imshow(grad_blur, 'gray')
```

```python
mask_im = (im + lap_im) * grad_blur
```

```python
plt.imshow(mask_im, 'gray')
```

```python
res_im = im + mask_im
```

```python
plt.imshow(res_im, 'gray')
```

```python
bright_im = np.sqrt(res_im)
```

```python
plt.imshow(bright_im, 'gray')
```

```python
#plt.subplot(121)
plt.figure(figsize=(7, 7))
plt.imshow(im, 'gray')
#plt.subplot(122)
plt.figure(figsize=(7, 7))
plt.imshow(bright_im, 'gray')
plt.show()
```

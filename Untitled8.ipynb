```
import sys
import cv2
import numpy as np
from scipy.ndimage import label
import matplotlib.pyplot as plt

def img_prero(img, orig):
  image = orig
  gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  ret,thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

  #plt.figure(num = 1, figsize=(size, size))
  #plt.subplot(121),plt.imshow(img)  
  fg = cv2.erode(thresh,None,iterations = 2)
  #plt.subplot(122),plt.imshow(fg)  
  bgt = cv2.dilate(thresh,None,iterations = 1)
  
  ret,bg = cv2.threshold(bgt,1,128,1)
  marker = cv2.add(fg,bg)
  marker_ = np.ones((292,299), dtype = complex)
  for i in range(image.shape[0]):
      for j in range(image.shape[1]):
          marker[i, j] = fg[i,j] + bg[i,j]
  
  plt.figure(num = 2, figsize=(size, size))
  plt.subplot(122),plt.imshow(np.dstack((marker, marker, marker)))  

  for i in range(image.shape[0]):
      for j in range(image.shape[1]):
        if marker[i, j] == 0:
          image[i, j] = [0] * 3 # [marker[i, j]]*3 

  plt.figure(num = 2, figsize=(size, size))
  plt.subplot(121),plt.imshow(image)  

  marker32 = np.int32(marker)
  cv2.watershed(img,marker32)
  m = cv2.convertScaleAbs(marker32)

  ret,thresh = cv2.threshold(m,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
  res = cv2.bitwise_and(img,img,mask = thresh)
  plt.subplot(122),plt.imshow(res)  
```

```
size = 20

orig = cv2.imread('1.png')
plt.figure(num = 1, figsize=(size, size))
plt.subplot(121),plt.imshow(orig)

```

```
img = cv2.bitwise_not(orig)
img = cv2.threshold(img,180,255, cv2.THRESH_TOZERO_INV)[1]
#plt.imshow()

img_prero(img, orig)
```

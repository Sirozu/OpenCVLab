```python
import cv2

import tkinter.filedialog as fd


import numpy as np

import matplotlib.pyplot as plt
```

```python
def select_image():
    path = "lena4.jpg" #fd.askopenfilename()
    if (len(path) > 0):
        return cv2.imread(path)
```

```python
def inc_H(image):
    hsv_im = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
    hsv_im[:,:,0] += 100
    return cv2.cvtColor(hsv_im, cv2.COLOR_YUV2BGR)
```

```python
def normalize(image):
    maxB = image[:,:,0].max()
    maxG = image[:,:,1].max()
    maxR = image[:,:,2].max()
         
    return np.uint8(image[:,:] * [255/maxB, 255/maxG, 255/maxR])
```

```python
def to_gray(image):    
    return image[:,:,2]*0.2126 + image[:,:,1]*0.7152 + image[:,:,0]*0.0772
```

```python
def from_gray(gray_image):
    image = np.zeros((gray_image.shape[0], gray_image.shape[1], 3), dtype=np.uint8)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            image[i,j] = np.uint8([np.round(gray_image[i,j])]*3)
            
    return image
```

```python
def hist(image):
    red = np.zeros(256)
    green = np.zeros(256)
    blue = np.zeros(256)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            blue[image[i,j,0]] += 1
            green[image[i,j,1]] += 1
            red[image[i,j,2]] += 1
            
    return (red/red.sum(), green/green.sum(), blue/blue.sum())
```

```python
def hist_gray(gray_im): 
    gray_hist = np.zeros(257)
    
    for i in range(gray_im.shape[0]):
        for j in range(gray_im.shape[1]):
            gray_hist[int(gray_im[i,j])] += 1
            
    gray_hist /= gray_im.shape[0]*gray_im.shape[1]
    
    return gray_hist
    
```

```python
def hist_comp(hist):
    comp_hist = np.zeros(256)
    
    for i in range(1,256):
        comp_hist[i] = comp_hist[i-1] + hist[i]
    
    return comp_hist
```

```python
def make_hist_image(hist, histW, histH, histColor):
    hist = hist/hist.max()
    binW = (int)(np.round(histW/256))
    histim = np.zeros((histH, binW*256, 3), dtype=np.uint8) + 255
    for i in range(256):
        for k in range(binW):
            for j in range(histH-1,(int)(np.round((1-hist[i])*histH)), -1):
                histim[j, binW*i + k] = histColor
            
    return histim
```

```python
def base_color(image, base_colr, targ_colr):
    blue = targ_colr[0]/(base_colr[0]+1)
    green = targ_colr[1]/(base_colr[1]+1)
    red = targ_colr[2]/(base_colr[2]+1)
                
    return normalize(image[:,:] * [blue, green, red])
```

```python
def gray_world(image):
    redAv = np.average(image[:,:,2])
    greenAv = np.average(image[:,:,1])
    blueAv = np.average(image[:,:,0])
    
    avg = (redAv + greenAv + blueAv)/3
    
    return normalize(image[:,:] * [avg/blueAv, avg/greenAv, avg/redAv])
```

```python
def log_correction(image):
    hsv_im = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
    hsv_im[:,:,0] = (np.log(hsv_im[:,:,0]))
    return cv2.cvtColor(hsv_im, cv2.COLOR_YUV2BGR)
```

```python
def pow_correction(image):
    hsv_im = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
    hsv_im[:,:,0] = (np.power(hsv_im[:,:,0], 2))
    return cv2.cvtColor(hsv_im, cv2.COLOR_YUV2BGR)
```

```python
def hist_norm(image):
    hsv_im = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
    #hsv_im[:,:,0] = (np.power(hsv_im[:,:,0], 2))
    histg = hist(image)[0]
    
    ymin = np.argmin(histg)
    ymax = np.argmax(histg)
    
    hsv_im[:,:,0] = (hsv_im[:,:,0] - ymin) * 255/(ymax - ymin)
    
    return cv2.cvtColor(hsv_im, cv2.COLOR_YUV2BGR)
```

```python
def hist_eq(gray_image):
    gray_hist = hist_gray(gray_image)
    comp_hist = hist_comp(gray_hist)
    
    for i in range(gray_image.shape[0]):
        for j in range(gray_image.shape[1]):
            gray_image[i,j] = 255*comp_hist[np.int(gray_image[i,j])]
            
    return gray_image
```

```python
def hist_eq_rgb(image):
    red_hist = hist_gray(image[:,:,2])
    green_hist = hist_gray(image[:,:,1])
    blue_hist = hist_gray(image[:,:,0])
    
    red_comp = hist_comp(red_hist)
    green_comp = hist_comp(green_hist)
    blue_comp = hist_comp(blue_hist)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            image[i,j] = np.uint8([255*blue_comp[image[i,j,0]], 255*green_comp[image[i,j,1]], 255*red_comp[image[i,j,2]]])
            
    return image                                                             
```

```python
def hist_test(image):
    red = np.zeros(256)
    red[image[:,:,2]] += 1
            
    return (red)
```

```python
def hand_bin(gray_image, low_bord, high_bord):
    for i in range(gray_image.shape[0]):
        for j in range(gray_image.shape[1]):
            if (gray_image[i,j] >= low_bord) and (gray_image[i,j] <= high_bord):
                gray_image[i,j] = 255
            else:
                gray_image[i,j] = 0
    
    return gray_image
```

```python
def otsu_bin(gray_image):
    hist = np.histogram(gray_image, normed=True, bins=256)[0] #hist_gray(gray_image)
    
    histint = np.cumsum(hist)
    histintr = np.cumsum(hist[::-1])
    
    i = np.arange(0, len(hist), 1)
    mu1 = np.cumsum(i*hist/histint)
    mu2 = np.cumsum(i[::-1]*hist[::-1]/histintr)[::-1]
    sigma = histint*(1-histint)*((mu1-mu2)**2)
    T = np.argmax(sigma)
    
    return hand_bin(gray_image, T, 255)
```

```python
def otsu_bin2(gray_image):
    hist = hist_gray(gray_image)
    
    q_old = hist[0]
    mu1 = 0
    mu2 = 0
    max_sigma = 0
    T = 0
    
    for t in range(1,255):
        q = q_old + hist[t]
        mu1 = (q_old*mu1 + t*hist[t])
        mu2 = (mu2 - q*mu1)/(1-q)
        
        sigma = q*(1-q)*((mu1 - mu2)**2)
        if (sigma > max_sigma):
            max_sigma = sigma
            T = t
        
    return hand_bin(gray_image, T, 255)
```

```python
def otsu_hier_quant(image, bord, low, k, kmax):
    mask = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if (not low and (image[i,j, 0] >= bord)) or (low and (image[i,j, 0] < bord)):
                mask[i,j] = 1;
    
    hist = cv2.calcHist([image], [0], mask, [256], [0,256]).reshape((-1,))
    hist = hist/hist.sum()
    
    for i in range(hist.shape[0]):
        if (hist[i] == 0):
            hist[i] = 1
    
    histint = np.cumsum(hist)
    histintr = np.cumsum(hist[::-1])
    
    i = np.arange(0, len(hist), 1)
    mu1 = np.cumsum(i*hist/histint)
    mu2 = np.cumsum(i[::-1]*hist[::-1]/histintr)[::-1]
    sigma = histint*(1-histint)*((mu1-mu2)**2)
    T = np.argmax(sigma)  
    print(k, low, T)
    
    res = mask*((k + kmax*low)*(image[:,:,0] > T));
    
    if (T > 0) and (T < 255) and (k<kmax):
        res = mask * (otsu_hier_quant(image, T, False, k+1, kmax) + otsu_hier_quant(image, T, True, k+1, kmax))
    
    return res
```

```python
def otsu_hier(image, bord, low, k, kmax):
    mask = np.zeros((image.shape[0], image.shape[1]), dtype=np.uint8)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if (not low and (image[i,j, 0] >= bord)) or (low and (image[i,j, 0] < bord)):
                mask[i,j] = 1;
    
    hist = cv2.calcHist([image], [0], mask, [256], [0,256]).reshape((-1,))
    hist = hist/hist.sum()
    
    for i in range(hist.shape[0]):
        if (hist[i] == 0):
            hist[i] = 1
    
    histint = np.cumsum(hist)
    histintr = np.cumsum(hist[::-1])
    
    i = np.arange(0, len(hist), 1)
    mu1 = np.cumsum(i*hist/histint)
    mu2 = np.cumsum(i[::-1]*hist[::-1]/histintr)[::-1]
    sigma = histint*(1-histint)*((mu1-mu2)**2)
    T = np.argmax(sigma)  
    print(k, low, T)
    
    res = mask*(image[:,:,0] > T);
    
    if (T > 0) and (T < 255) and (k<kmax):
        res = mask * (otsu_hier(image, T, False, k+1, kmax) + otsu_hier(image, T, True, k+1, kmax))
    
    return res
```
